```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("paysim_sample.csv")
df.head()
```

```python
df.shape
```

```python
#basic cleaning of the dara

df=df.drop_duplicates()
df=df.reset_index(drop=True)
```

```python
df.shape  #no duplicates present
```

```python
#dropping unnecessary columns

df=df.drop(columns=['nameOrig','nameDest'])
df.head()
```

```python
df.dtypes
```

```python
# feature engineering

#for anomolies and payment frauds - oldbalanceOrg / newbalanceOrig -Balance anomalies

#oldbalanceOrg − amount ≠ newbalanceOrig, so now here we calculate the differnce, if the it doesnt match its an anomoly

df["balance_delta_org"] = df["oldbalanceOrg"] - df["newbalanceOrig"]
df["balance_delta_dest"] = df["newbalanceDest"] - df["oldbalanceDest"]
```

```python
df["hour"] = df["step"] % 24
```

```python
df = pd.get_dummies(df, columns=["type"], drop_first=True)
```
